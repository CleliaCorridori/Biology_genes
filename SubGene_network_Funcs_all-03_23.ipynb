## MATRICES OF THE INTERACTIONS 

### With Spectral Decomposition

```python
# np.fill_diagonal(C_Mean, 0)

THR = 0.001

eigval, eigvect = funcs.eigval_analysis(C_Mean, thr = THR, info=False, fig=False)
# print(eigval[-2:])


# check
print("original determinant", la.det(np.diag(eigval)))
print("determinant with cut eigenvalues", la.det(np.diag(eigval[:15])))
print(eigval[14])
```

```python
#pseudo inverse
Mint = -np.linalg.pinv(C_Mean, rcond=0.00001/np.max(eigval))
# print("check (differences between pseudo-inverse and inverse matrices' elements)", len(Mint[Mint-la.inv(C_Mean)>0.001]))
# print("check (differences between pseudo-inverse and inverse matrices' elements)", len(Mint[Mint-MintP>0.001]))

# figures 
perc = 0.1
level = np.max(Mint)*perc #5%
print("level of the cut-off", level, str(perc*100)+"% of ", np.max(Mint))
funcs.fig_matx_int(Mint, genes, "all", thr = level, lintsh = 0.1)

# Info
rows, cols = funcs.ind_val_abs(Mint, 10, False)
gene_names = funcs.gene_list(rows, cols, genes)
print("The ten strongest interactions: \n", gene_names)

```

```python
# distribution of interaction coefficients
int_vect = np.squeeze(np.reshape(Mint, (np.shape(Mint)[0]*np.shape(Mint)[1],1)))
# int_vect_2i = np.squeeze(np.reshape(Interac_matx_2i, (np.shape(Interac_matx_2i)[0]*np.shape(Interac_matx_2i)[1],1)))
print(np.shape(Mint),np.shape(int_vect))
plt.figure(figsize=(12,6))
plt.hist(int_vect, 30, density=False, color="Navy")
plt.grid()
plt.title("2i+LIF", fontsize=14)
plt.xlabel("Interaction Coefficients", fontsize=14)
plt.ylabel("Counts", fontsize=14)
thresh = level
plt.axvline(thresh, color="red", label="+"+np.str(np.round(thresh,2) ))
plt.axvline(-thresh, color="red", label="-"+np.str(np.round(thresh,2) ))
plt.legend(fontsize="14")
# plt.xscale("log")


```
